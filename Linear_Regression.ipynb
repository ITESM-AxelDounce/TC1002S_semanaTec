 # Polynomial Linear Regression

Importamos las librerías de sklearn:
* Empezamos con las funciones de regresión lineal

```python
from sklearn.linear_model import LinearRegression
```

Aquí importamos otras bibliotecas pertinentes que ya se han utilizado con anterioridad.

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

El modelo general con el se trata de ajustar los datos es el siguiente:
y = b<sub>0</sub> + b<sub>1</sub>x<sub>1</sub> + b<sub>2</sub>x<sub>2</sub> ...

En este caso vamos a ajustar un modelo de grado uno, es decir, el modelo de una línea recta.

## Cargamos datos

Cargamos datos con pandas en formato CSV, y luego definimos nuestras variables con las que queremos encontrar una relación, este caso "Peso" contra "Estatura". Para esto es necesario convertir a arreglos numpy. Y también convertir la variable "X" a dos dimensiones.

```python
df = pd.read_csv("peso_estatura_genero.csv")
x= df["Peso"].to_numpy().reshape((-1,1))
y = df["Estatura"].to_numpy()
```

## El modelo

Creamos el modelo de la Regresión Lineal:

```python
model = LinearRegression()
```

Agregamos las variables al modelo:

```python
model.fit(x,y)
```

Depués podemos observar nuestra constante calculada del modelo *b<sub>0</sub>*:

```python
model.intercept_
```

Y lo mismo para nuestra constante calculada del modelo *b<sub>1</sub>*:

```python
model.coef_
```

Finalmente calculamos la nueva variable ajustada "Y":

```python
y2 = model.predict(x)
```

Observamos que nuestro modelo esté adecuadamente ajustado:

```python
model.score(x,y)
```

Al final graficamos para observar los datos y el modelo:

```python
plt.scatter(x,y)
plt.plot(x,y2,"r")
plt.show()
```

## Usando seaborn

Alternativamente, en lugar de usar la biblioteca sklearn, podemos usar directamente seaborn:

```python
sns.lmplot(data=df,x="Peso",y="Estatura",col="Genero",line_kws={"color":"red"})
```
